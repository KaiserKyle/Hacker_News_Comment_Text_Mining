```python
import nltk # be sure to have stopwords installed for this using nltk.download_shell()
import pandas as pd 
import string

data = pd.read_csv("d:\\data\\ISTFinal\FinalProjectData.csv")

from nltk.sentiment.vader import SentimentIntensityAnalyzer
# install Vader and make sure you download the lexicon as well
sid = SentimentIntensityAnalyzer()
# this step will return an error if you have not installed the lexicon

data.columns

posdata = data.loc[data['GoodComment'] == True]
negdata = data.loc[data['GoodComment'] == False]
```

```python
possummary = {"positive":0,"neutral":0,"negative":0}
vaderResults = []
for x in data['CommentText']: 
    #print(x)
    ss = sid.polarity_scores(str(x))
    vaderResults.append({"compound": ss['compound'], "pos": ss['pos'], "neg": ss['neg'], "neu" : ss['neu']})
    if ss["compound"] == 0.0: 
        possummary["neutral"] +=1
    elif ss["compound"] > 0.0:
        possummary["positive"] +=1
    else:
        possummary["negative"] +=1
        
df = pd.DataFrame(vaderResults)
df.shape
```

```python
outputDf = pd.concat([data,df], axis = 1)
outputDf.head
```

```python
outputDf.to_csv("d:\\data\\ISTFinal\\FinalProjectDataSentiment.csv")
```
